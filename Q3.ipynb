```python
input_file = open("ML_Input.txt", "r")
input = input_file.read()
input_file.close()
```

```python
print(input)
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopwords= list(STOP_WORDS)
other_sw= ['sir','afternoon','morning','thank','thankyou','ahh','ah','best','me','yes','tell','okay','!',',','.','?','@',':',';','-','"','\n','\n \n']
stopwords = stopwords + other_sw
print(stopwords)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(input)
```

```python
tokens = [token.text.lower() for token in doc]
```

```python
# punctuation =punctuation + '/n'
```

```python
word_freq = {}
for wd in doc:
    if wd.text.lower() not in stopwords:
        if wd.text.lower() not in punctuation:
            if wd.text not in word_freq.keys():
                word_freq[wd.text]=1
            else:
                word_freq[wd.text] += 1
```

```python

print(word_freq)
```

```python
max_freq = max(word_freq.values())
max_freq
```

```python
for word in word_freq.keys():
    word_freq[word] = word_freq[word]/max_freq
print(word_freq)
```

```python
sent_tokens = [sent for sent in doc.sents]
print((sent_tokens))
```

```python
sentence_scores = {}
for sent in sent_tokens:
    for word in sent:
        if word.text.lower() in word_freq.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent]= word_freq[word.text.lower()]
            else:
                sentence_scores[sent] += word_freq[word.text.lower()]
                
```

```python
sentence_scores
```

```python
from heapq import nlargest
```

```python
select_length = int(len(sent_tokens)*0.15)
select_length
```

```python
summary = nlargest(select_length, sentence_scores, key= sentence_scores.get)
summary
```

```python
final_summary0 = [word.text for word in summary]
str =" "
final_summary= str.join(final_summary0)
print(final_summary)
```
